```python
#parsing energy use through the world bank
import pandas as pd
df = pd.read_excel('energy.xls', skiprows=[0,1, 2])
df = df[['Country_Name','2014']] #2014 based
df
```

```python
#filters list by developing country by world bank
developed = ['Andorra', 'Australia', 'Austria', 'Belgium', 'Bermuda', 'Canada', 'Denmark', 'Faroe Islands', 'Finland', 'France', 'Germany', 'Greece', 'Holy See', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan', 'Liechtenstein', 'Luxembourg', 'Malta', 'Monaco', 'Netherlands', 'NZ', 'Norway', 'Portugal', 'San Marino', 'South Africa', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 'United States']
dfdev = df[df['Country_Name'].isin(developed)]
dfdev = dfdev.dropna(thresh=2)

dfdev
```

```python
# average kg oil per capita
from decimal import Decimal
def scino(x):
    return "{:.2e}".format(Decimal(str(x)))

oilcapita = sum(dfdev['2014'])/len(dfdev)
oilcapita_adj = oilcapita * 0.15
oilcapita_adj_4bil = oilcapita_adj * 4e9

print("4 billion using 15%")
print(scino(oilcapita_adj_4bil))
```

```python
#calculates value for the US total usage
us_oil_capita = 6955.52
us_pop2014 = 3.16e8

us_oil = us_oil_capita * us_pop2014



# calculating for 4bil
print("US Total Energy")
print(scino(us_oil))
```
